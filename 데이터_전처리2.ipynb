```python
# 50~59세의 신장평균을 구하여라

df['측정나이'] = df['측정나이'].astype('int')
df['나이대'] = df['측정나이'].map(lambda x: x//10*10 )
df.loc[df['나이대']==50].iloc[:,2].mean()
```

```python
# 남성 중 A등급과 D등급의 체지방률 평균의 차이(큰 값에서 작은 값의 차)를 구하여라

dd = df.loc[df['측정회원성별'].str.contains('M')]
A = dd.loc[dd['등급'].str.contains('A')].iloc[:,4].mean()
D = dd.loc[dd['등급'].str.contains('D')].iloc[:,4].mean()
D-A
```

```python
# bmi는 자신의 몸무게(kg)를 키의 제곱(m)으로 나눈값이다. 데이터의 bmi 를 구한 새로운 컬럼을 만들고 남성의 bmi 평균을 구하여라

df['bmi'] = df['체중 : kg']/(df['신장 : cm']/100)**2
df.loc[df['측정회원성별'].str.contains('M')].bmi.mean()
```

```python
# bmi보다 체지방율이 높은 사람들의 체중평균을 구하여라

df.loc[df['bmi'] <= df['체지방율 : %']].iloc[:,3].mean()
```

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder as le
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler

from sklearn.model_selection import train_test_split

from sklearn.metrics import roc_auc_score, f1_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
def get_scores(model, X_train, X_test, y_train, y_test):
  A = model.score(X_train, y_train)
  B = model.score(X_test, y_test)
  y_pred = model.predict(X_test)
  C = f1_score(y_test, y_pred)
  y_proba = model.predict_proba(X_test)[:,1]
  D = roc_auc_score(y_test, y_proba)
  return f'accuracy : {A:.4f} {B:.4f}     f1_score : {C:.4f}     roc_auc_score : {D:.4f}'
```

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder as le
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler

from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_percentage_error as mape
from sklearn.metrics import mean_squared_log_error as msle

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
def get_scores(model, X_train, X_test, y_train, y_test):
  A = model.score(X_train, y_train)
  B = model.score(X_test, y_test)
  y_pred = model.predict(X_test)
  y_pred = np.where(y_pred>0, y_pred, -y_pred)
  C = mse(y_test, y_pred)**0.5
  D = mape(y_test, y_pred)
  E = msle(y_test, y_pred)**0.5
  return f'R2 : {A:.4f}  {B:.4f}     rmse : {C:.4f}    mape : {D:.4f}      rmsle : {E:.4f}'
```
