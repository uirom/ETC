```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/경기도 화성시_경기도 화성시 동탄신도시 도로별 통행량 및 통행로 정보_20211105.csv",encoding='utf-8')
```

```python
df.head()
```

# 일을 기준으로 교통량(대)의 합을 알아보기

```python
df = df.drop(columns=['길이'])
```

```python
df['일자'] = pd.to_datetime(df['일자'])
df[['일자','교통량(대)']].groupby('일자').sum()
```

# 몇시간 단위로 체크된 데이터인지 시간 데이터 이용하여 찾아보기. 그리고 그 단위를 컬럼에 추가

```python
df['시작시간'] =df['시간'].str.split('~').str[0]
df['시작시간'] = pd.to_datetime(df['시작시간'])
df['끝나는시간'] = df['시간'].str.split('~').str[1]
df['끝나는시간'] = pd.to_datetime(df['끝나는시간'])
```

```python
df['endtime'] = df['끝나는시간'].dt.hour
df['starttime'] = df['시작시간'].dt.hour
```

```python
df['시간대'] = df['endtime']-df['starttime']
df
```

# df['소통정보(소통판정)'] 이용하여 정체, 서행, 원할이 몇개인지 찾아보기

```python
df['소통정보(소통판정)']. value_counts()
```

# 제일 교통량이 많은 곳 찾아보기 (구간정보와 소재지 주소를 출력)

```python
df.sort_values('교통량(대)', ascending=False).reset_index(drop=True)[['소재지 주소','구간정보']].iloc[0]
```

# 구간정보중 '삼거리'가 들어간 데이터들의 속도 평균을 구해보자

```python
dd = df.loc[df['구간정보'].str.contains('삼거리')]
dd['속도(km)'].mean()
```

# '소재지 주소' 이용하여 동별로 교통량(대) 총 수 구하기

```python
df['소재지 주소'].value_counts()
```

```python
df[['소재지 주소','교통량(대)']].groupby('소재지 주소').sum().sort_values('교통량(대)', ascending=False)
# 화성시 능동 1081-1이 제일 교통량(대)가 많다.
```

# 소재지 주소 이용하여 동별 합산속도가 제일 높은 구간 구하기

```python
df[['소재지 주소','속도(km)']].groupby('소재지 주소').sum().sort_values('속도(km)',ascending=False).reset_index().iloc[0,0]
```
